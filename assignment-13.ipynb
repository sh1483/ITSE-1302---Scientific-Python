```python
# Plot the hyperbolic tangent (tanh()) function using six different color formats.
# Modify x slightly in each plot so the lines are distinct.

plt.plot(x, np.tanh(x), color='green')        # specify color by name
plt.plot(x, np.tanh(x - 1), color='b')           # short color code (rgbcmyk)
plt.plot(x, np.tanh(x + 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.tanh(x * 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.tanh(x ** 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.tanh(x / 5), color='chartreuse'); # all HTML color names supported
```

## Requirement 17

```python
# Produce the following plot (purple and orange).
# Create the function we will use
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
# Using np.meshgrid to for a 2D grid
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# Combining a contour plot and image plot
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='PuOr', alpha=0.5)
plt.colorbar();
```

## Requirement 19

```python
# Produce the following plot (notice the color and legends).

fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='olivedrab') # Not right, but close enough on color
ax.axis('equal')

# make the top legend
ax.legend(lines[:2], ['line A', 'line B'],
          loc='upper left', frameon=True, fancybox=False)

# make the bottom legend
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower left', frameon=True, fancybox=False)
ax.add_artist(leg); # add the artist
```
